```python
df['total_spend'] = df['TV Ad Budget ($)']+df['Radio Ad Budget ($)']+df['Newspaper Ad Budget ($)']
df.head()
```

```python
df.drop(columns=['TV Ad Budget ($)','Radio Ad Budget ($)','Newspaper Ad Budget ($)'], inplace=True)
df.head()
```

```python
df.describe()
```

```python
sns.pairplot(df)
plt.show()
```

```python
X = df[['total_spend']]
y= df['Sales ($)']
```

```python

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                    test_size = 0.3,
                                                    random_state = 142)
```

```python
from sklearn.linear_model import LinearRegression # import the model

# save the model
LR_model = LinearRegression()

#fit

LR_model.fit(X_train, y_train)
```

```python


y_pred = LR_model.predict(X_test)
y_pred
acc = LR_model.score(X_test, y_test)
print(acc)
```

```python
plt.figure(figsize=(10, 6))

# Plot actual sales data in blue
plt.plot(np.arange(len(y_test)), y_test, marker='o', linestyle='-', color='blue', label='Actual Sales')

# Plot predicted sales data in red
plt.plot(np.arange(len(y_test)), y_pred, marker='o', linestyle='-', color='red', label='Predicted Sales')

plt.title('Time Series Forecasting of Sales')
plt.xlabel('Time')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```
